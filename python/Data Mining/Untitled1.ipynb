```python
dfs = []
files = ['Humidity.csv', 'Temperature.csv', 'Speed.csv', 'PM25.csv', 'PM10.csv', 'PM1.csv', 'NO2.csv', 'BC.csv']
for file in files:
    dfs.append(pd.read_csv(file, header=None))

i=0
for df in dfs:
    if(i == 0):
        for i in range(10,107089):
            df[1][1] = i%5
    else:
        for i in range(0,108330):
            df[1][1] = i%5
```

```python
for df in dfs:
    df.set_index([0,1], drop = True, inplace=True)
```

```python
Data = reduce(lambda df_left, df_right:
              pd.merge(df_left, df_right, left_index = True, right_index=True, how='inner'), dfs).fillna(np.nan)
```
