```python
#ici j'importe les bibliotheques que j'ai besoi pour mon tp
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#chargement des donn√©es d'entrainement
house_data = pd.read_csv('house.csv')
house_data = house_data[house_data['loyer'] < 10000]
```

```python
#Affichage de notre nuange de donn√©es
plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)
plt.show()
```

```python
x = np.matrix([np.ones(house_data.shape[0]), house_data['surface']]).T
x[:10]
y = np.matrix([house_data['loyer']]).T
y[:10]
```

```python
theta = np.linalg.inv(x.T.dot(x)).dot(x.T).dot(y)
print(theta)
```

```python
#ici je peux faire une prediction avec un appartement de surface 30 m2  
k = theta.item(0) + theta.item(1) * 35
print(k)
```

```python
plt.xlabel('Surface')
plt.ylabel('Loyer')

plt.plot(house_data['surface'], house_data['loyer'], 'ro', markersize=4)

# On affiche la droite entre 0 et 250
plt.plot([0,250], [theta.item(0),theta.item(0) + 250 * theta.item(1)], linestyle='-', c='#000000')

plt.show()
```
